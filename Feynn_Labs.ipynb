<h3><font color='red'>Hierarchical Clustering</font></h3>

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage
```

```python
Z=linkage(data,method='single',metric='euclidean',optimal_ordering=False)
```

```python
plt.figure(figsize=(15,10))
dendrogram(Z,p=30)
```

```python
Z_ward=linkage(data,method='ward',metric='euclidean',optimal_ordering=False)
plt.figure(figsize=(15,10))
dendrogram(Z_ward,p=30)
```

```python
Agg_clustering=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage="ward")
Agg_clustering.fit_predict(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data['Age'],data['VisitFrequency'],c=Agg_clustering.labels_,cmap='rainbow')
```

<h3><font color='red'>K-means Clustering</font></h3>

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
model=KMeans()
visualizer=KElbowVisualizer(model,k=(2,20),metric="distortion")
visualizer.fit(data)
visualizer.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(2)
df=pca.fit_transform(data)
```

```python
print(df.shape)
print(df[:5,:])
```

```python
kmeans=KMeans(n_clusters=5,random_state=1).fit(df)
centroids=kmeans.cluster_centers_
```

```python
label=kmeans.fit_predict(df)
print(label)
```

```python
np.unique(label)
```

```python
for i in range(0,5):
  data_points=df[label==i]
  plt.scatter(data_points[:,0],data_points[:,1])
plt.scatter(centroids[:,0],centroids[:,1],color='k')
plt.show()
```

<h3><font color='red'>Hybrid Approach</font></h3>
<h4><font color='green'>KMeans+Agglomerative Clustering</font></h4>

```python
kmeans=KMeans(n_clusters=100,random_state=1).fit(df)
centroids=kmeans.cluster_centers_
Label=kmeans.fit_predict(df)
print(Label)
```

```python
np.unique(Label)
```

```python
Centroid_points=pd.DataFrame(centroids)
```

```python
Centroid_points.shape
```

```python
Z_ward=linkage(Centroid_points,method='ward',metric='euclidean',optimal_ordering=False)
plt.figure(figsize=(15,10))
dendrogram(Z_ward,p=30)
```

```python
Agg_clustering=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage="ward")
final_Label=Agg_clustering.fit_predict(Centroid_points)
```
